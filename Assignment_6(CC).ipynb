```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

M = int(input())
x = np.linspace(-10, 10)
def y(x):
    return M * (x**2)

print(y(x))

def Y(x):
    return M * (np.sin(x))
print(Y(x))

plt.figure(figsize=(8, 6))
plt.plot(x, y(x), label='y = M * x^2', color='blue', linestyle='-')
plt.plot(x, Y(x), label='y = M * sin(x)', color='red', linestyle='--')

plt.legend()
plt.grid(True)
plt.title('Plot of both functions')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
data ={"Subjects": ["Eng", "Math", "Sci", "S.S.T", " Hindi"],
        "Scores": [88, 100, 58, 20, 89]}
df=pd.DataFrame(data)
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Subjects", y="Scores", data=data, palette="rocket")

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.title("Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(True)

plt.show()
```

```python
rno = int(input())
np.random.seed(rno)
dataset = np.random.randn(50)
fig, axes = plt.subplots(2,2,figsize = (10,8))

axes[0, 0].plot(np.cumsum(dataset),marker ="o", linestyle="-")
axes[0, 0].set_title('Cumulative Sum')
axes[0, 0].set_xlabel('Data Point')
axes[0, 0].set_ylabel('Cumulative Sum')
axes[0, 0].grid(True)

noise = np.random.randn(50) * 0.2
axes[1, 1].scatter(dataset, dataset + noise, marker='x', color='yellow')
axes[1, 1].set_title('Scatter Plot with Random Noise')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Y-axis')
axes[1, 1].grid(True)

plt.show()
```
